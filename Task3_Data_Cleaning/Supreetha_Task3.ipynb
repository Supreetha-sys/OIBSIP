# Data Cleaning using Python

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load the Dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df.head()
```

## 3. Dataset Shape

```python
df.shape
```

## 4. Dataset Information

```python
df.info()
```

## 5. Missing Values

```python
df.isnull().sum()
```

## 6. Fill Missing Values in Age

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Age"]
```

```python
df["Age"].isnull().sum()
```

## 7. Fill Missing Values in Embarked

```python
df["Embarked"].isnull().sum()
```

## 8. Remove Cabin Column

```python
df.drop("Cabin", axis=1, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

## 9. Check Duplicate Rows

```python
df.duplicated().sum()
```

## 10. Save the Cleaned Dataset

```python
df.to_csv("Cleaned_Titanic_Dataset.csv", index=False)
```

```python
print("Dataset saved successfully!")
```

# Observations

- The dataset contains 891 rows and 12 columns.
- The Age column had 177 missing values, which were filled using the median.
- The Embarked column had 2 missing values, which were filled using the mode.
- The Cabin column had 687 missing values (about 77%), so it was removed.
- No duplicate rows were found in the dataset.
- The cleaned dataset was successfully saved as a new CSV file.

# Conclusion

The Titanic dataset was successfully cleaned by handling missing values, removing the Cabin column due to excessive missing data, checking for duplicates, and saving the cleaned dataset. The dataset is now ready for further analysis or machine learning.
